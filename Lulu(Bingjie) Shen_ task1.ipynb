```python
import numpy as np;
import pandas as pd;
df_origin= pd.read_csv("Lenovo Dataset_Lenovo Dataset.csv",encoding='ISO-8859-1')
df_origin.head()
```

```python
# Based on gender and whether he or she is in a small business, segmenting the audience into 4 parts.
# Respectively, they are Male in small business, female in small business, male in very small business and female in
# very small business. 
df = [0] * 4
df[0] = df_origin[(df_origin['SEGMENT_CD'] == 'SMB') & (df_origin['gender'] == 'Male')]
df[1] = df_origin[(df_origin['SEGMENT_CD'] == 'VSB') & (df_origin['gender'] == 'Male')]
df[2] = df_origin[(df_origin['SEGMENT_CD'] == 'SMB') & (df_origin['gender'] == 'Female')]
df[3] = df_origin[(df_origin['SEGMENT_CD'] == 'VSB') & (df_origin['gender'] == 'Female')]
df
```

```python
states = df_origin['MAIL_REGION'].unique()
states
```

```python
total = [[] for i in range(4)]
quant = [[] for i in range(4)]
avg = [[] for i in range(4)]
for i in range(4):
    for state in states:
        temp = df[i][df[i]['MAIL_REGION'] == state]
        total[i].append(temp['REVENUE'].sum())
        quant[i].append(temp['QUANTITY'].sum())
for i in range(4):
    for j in range(len(states)):
        avg[i].append(total[i][j] * 1.0 / quant[i][j])
cols = ['Male-SMB', 'Male-VSB', 'Female-SMB', 'Female-VSB']
rv = pd.DataFrame(index=states, columns=cols)
for i in range(4):
    for j in range(len(states)):
        rv[cols[i]][states[j]] = avg[i][j]
rv
```

```python
for m in range(4):
    for state in states:
        temp1 = df[m][df[m]['MAIL_REGION'] == state]
        total[m].append(temp1['REVENUE'].sum())
for m in range(4):
    for n in range(len(states)):
        total[m].append(total[m][n])
cols = ['Male-SMB', 'Male-VSB', 'Female-SMB', 'Female-VSB']
rv_total = pd.DataFrame(index=states, columns=cols)
for m in range(4):
    for n in range(len(states)):
        rv_total[cols[m]][states[n]] = total[m][n]
rv_total
```

```python
export_excel = rv.to_excel ('revenue distribution table.xlsx', index = True, header=True)
```

```python
export_excel = rv_total.to_excel ('revenue_total distribution table.xlsx', index = True, header=True)
```
